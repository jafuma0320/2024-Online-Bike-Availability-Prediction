En el siguiente paso, lo que hice fue tratar de unificar los codigos anteriormente con el ejemplo que he tomado.

```python
def compute_year(sample_ts):
  dt_object_utc = datetime.fromtimestamp(sample_ts, pytz.UTC)
  dt_object_spain = dt_object_utc.astimezone(pytz.timezone('Europe/Madrid'))
  return dt_object_spain.year
```

```python
df_example['year'] = df_example['last_reported'].apply(compute_year)
```

Aqui lo que se puede observar es que se ha agregado una ultima columna llamada Year, esto se realizo como por ejemplo y comprobacion.

```python
df_example
```

Como observe que el codigo funciono bien para el ano, ahora voy a realizar la misma tarea para el mes, el dia y la hora.

```python
def compute_month(sample_ts):
  dt_object_utc = datetime.fromtimestamp(sample_ts, pytz.UTC)
  dt_object_spain = dt_object_utc.astimezone(pytz.timezone('Europe/Madrid'))
  return dt_object_spain.month
df_example['month'] = df_example['last_reported'].apply(compute_month)
```

```python
def compute_day(sample_ts):
  dt_object_utc = datetime.fromtimestamp(sample_ts, pytz.UTC)
  dt_object_spain = dt_object_utc.astimezone(pytz.timezone('Europe/Madrid'))
  return dt_object_spain.day
df_example['day'] = df_example['last_reported'].apply(compute_day)
```

```python
def compute_hour(sample_ts):
  dt_object_utc = datetime.fromtimestamp(sample_ts, pytz.UTC)
  dt_object_spain = dt_object_utc.astimezone(pytz.timezone('Europe/Madrid'))
  return dt_object_spain.hour
df_example['hour'] = df_example['last_reported'].apply(compute_hour)
```

```python
# Define la ruta de la carpeta
folder_path ='./'

# Liste todo los archivos (CSV)
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

### Voy a crear un diccionario para almacenar los dataframes

```python
# Diccionario para almacener Dataframes
dataframes = {}

# Loop (recorrer todos los archivos CSV y tratar de leer cada uno de ellos con un dataframe)
for file in csv_files :                          # Bucle que recorre todos los archivos csv
    file_path = os.path.join(folder_path, file)  # Define la ubicación del archivo csv uniendo el nombre del archivo con el path (directorio actual, ./)
    df = pd.read_csv(file_path, sep = ',')       # Leer en un dataframe (tipo de datos de la librería pandas) el archivo csv que se está procesando en la iteración actual

    df['year'] = df['last_reported'].dropna().apply(compute_year) # Adiccione columna con el año # El dropna() elimina cualquier valur nulo(vacio)en la columna 'last_reported'
    df['month'] = df['last_reported'].dropna().apply(compute_month) # Adiccione columna con el mes
    df['day'] = df ['last_reported'].dropna().apply(compute_day) # Adiccione columna con el dia
    df['hour'] = df['last_reported'].dropna().apply(compute_hour) # Adiccione columna con el hora

    df_simplified = df[['station_id', 'num_docks_available', 'num_bikes_available', 'year','month','day','hour']] # Elimine columnas que no voy a usar y cree un DataFrame simplificado del DataFrame original df

    grouped_series = df_simplified.groupby(['station_id','year', 'month','day','hour']).mean() # realice una agrupacion por estación, año, mes, día y hora, y saque la media a los muelles disponibles en cada hora
    grouped_series = grouped_series[['num_docks_available','num_bikes_available']]
    grouped_series = grouped_series.reset_index()

    dataframes[file] = grouped_series
```

```python
# cONCATENAR todos los archivos processasod en u nunico dataframe, y guardarlos en un archivo CSV en drive
all_data = pd.concat(dataframes.values(), ignore_index=False)
all_data.to_csv('/content/drive/MyDrive/University of Barcelona/Capstone_Project_2024/combined_data_2.csv', index=False)
```
