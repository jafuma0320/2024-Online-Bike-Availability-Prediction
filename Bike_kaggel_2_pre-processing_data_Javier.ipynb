```python
def compute_proportion_docks(station_n, num_docks):
    # Filtrar las filas del DataFrame para obtener las filas con el station_id especificado
    filtered_df = df_info_simplified[df_info_simplified['station_id'] == station_n]

    # Imprimir las filas filtradas para depuración
    print(filtered_df)

    # Verificar si el DataFrame filtrado está vacío
    if filtered_df.empty:
        # Manejar el caso donde no se encuentra la estación
        print(f"No se encontró la estación con station_id: {station_n}")
        return None  # O algún valor predeterminado que tenga sentido en tu contexto

    # Obtener la capacidad de la primera fila del DataFrame filtrado
    capacity = filtered_df['capacity'].iloc[0]

    # Calcular la proporción de muelles disponibles
    prop_docks = num_docks / capacity

    return prop_docks

# Aplicar la función a cada fila del DataFrame df y crear una nueva columna
df['proportion_docks_available'] = df.apply(lambda row: compute_proportion_docks(row['station_id'], row['num_docks_available']), axis=1)

# Mostrar las primeras filas del DataFrame para verificar el resultado
df.head()
```
