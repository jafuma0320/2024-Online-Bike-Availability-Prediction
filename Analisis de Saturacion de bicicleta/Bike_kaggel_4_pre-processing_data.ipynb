Voy a aplicar este cálculo a todas las filas de mi Dataframe

```python
# Function to get the weekday
def get_weekday(row):
    date = datetime.date(int(row['year']), int(row['month']), int(row['day']))
    return date.weekday()

# Apply the function to each row and create a new column
result_df['weekday'] = result_df.apply(lambda row: get_weekday(row), axis=1)
result_df
```

Agregue una nueva columna con información sobre si es Fin de de Semana

```python
# Function to check if a date is a weekend
def is_weekened(row):
    weekday = row['weekday']
    if weekday == 5 or weekday == 6:
      return True
    return False

# Apply the function to each row and create a new column
result_df['is_weekend'] = result_df.apply(lambda row: is_weekened(row), axis=1)
result_df
```

Agregue una nueva columna con información sobre si es Festivo

```python
# Function to check if a date is a holiday
def is_holiday(row):
    date = datetime.date(row['year'], row['month'], row['day'])

    # Initialize the holidays library for the given country, spain ES
    holiday_list = holidays.ES()

    return date in holiday_list

# Apply the function to each row and create a new column
result_df['is_festive'] = result_df.apply(lambda row: is_holiday(row), axis=1)
result_df
```

## 8.  Analisis mapa con información del porcentaje de muelles disponibles de los Festivos o Fin de Semana, entre 12h00 & 23h00

Primero filtre los datos por Festivos o Fin de Semana y en el horario entre las 12h00 & 23h00.

Después saque/identifique/genere la media del valor de la proporción de muelles disponibles en ese tramo horario para cada estación.

```python
result_df_2023_filtered = result_df_2023[((result_df_2023['is_festive'] == True) | (result_df_2023['is_weekend'] == True)) & (result_df_2023['hour'] > 12) & (result_df_2023['hour'] < 23)]
result_df_2023_filtered = result_df_2023_filtered[['station_id', 'percentage_docks_available']].groupby('station_id').mean()
result_df_2023_filtered = result_df_2023_filtered.reset_index()
result_df_2023_filtered
```

Pege/integre la media de la proporción de muelles disponibles con la información de las estaciones para hacer el mapa.

```python
df_filtered = df_info[df_info['station_id'].isin(list(result_df_2023_filtered['station_id'].values))]
```

```python
df_merged = pd.merge(df_filtered, result_df_2023_filtered, on='station_id', how='left')

df_merged.head()
```

# 9.  Analisis y generacion del mapa con el tamaño de los círculos en función del Inverso del percetage_docks_available.

values = 10*(1-df_merged['percentage_docks_available'][:]) : es la proporción de bicicletas disponibles en la estación.

```python
lats = df_merged['lat'][:]
lons = df_merged['lon'][:]
post_codes = df_merged['post_code'][:]
cities = ['Barcelona'] * len(lats)
values = 10*(1-df_merged['percentage_docks_available'][:])
ids = df_merged['station_id'][:]

# Step 1: Create synthetic data for postal codes
data = {
    'postal_code': post_codes,
    'city': cities,
    'value': values,
    'latitude': lats,
    'longitude': lons,
    'ids': ids
}

df = pd.DataFrame(data)

# Step 2: Create a folium map centered around Barcelona
m = folium.Map(location=[41.39, 2.18], zoom_start=12)

# Step 3: Add markers to the map for each postal code
for _, row in df.iterrows():
    folium.CircleMarker(
        location=(row['latitude'], row['longitude']),
        radius=row['value'],
        color='blue',
        fill=True,
        fill_color='blue',
        popup=f"Postal Code: {row['postal_code']}<br>City: {row['city']}<br>Value: {row['value']}<br>Id: {row['ids']}"
    ).add_to(m)

# Display the map
m
```

Este último mapa muestra las estaciones según el nivel de saturación durante el día entre 12h00 & 23h00 de días festivos y fines de semanas.

Las estaciones con más saturación tienen círculos más grandes, es decir, aquellas que tienen más bicicletas según capacidad y por tanto menos muelles disponibles.
