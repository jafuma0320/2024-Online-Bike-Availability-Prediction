Lo siguiente es calcular los contextos de mis datos (CXT-4, CXT-3, CXT-2, CXT-1) estos son los contextos.
Con los anteriores contextos tengo que coger agrupaciones de 5 filas seguidas, segun instrucciones  de PERE. "Corresponderian a 5 horas seguidas, de informacion sobre/en una estacion"
***Las filas las voy a tomar de 5 en 5 por station_id de c/d estacion, por eso hay que segmentar las estaciones primero.

Calcule los contextos para todas las estaciones y sus datos. Para ello primero segmente por cada estación y cree una lista con todas las estaciones disponibles

```python
# Lista de estaciones
counter_ids_filtered = Counter(df_ordenado['station_id'])
stations_filtered = list(counter_ids_filtered.keys())
stations_filtered
```

```python
# Comprobe que no me hace falta ninguna estacion
len(stations_filtered)
```

En el primer Bucle recorri todas las estaciones.    
En el segundo Bucle, recorri todas las filas de 5 en 5 de una estación concreta para calcular todos los contextos.

```python
result_list = []
for station_id in stations_filtered:  # Bucle, en cada iteracion se procesa 1 station_id
  print(f'Analyzing station {station_id}')
  df_n_station = df_ordenado[df_ordenado['station_id'] == station_id] # Cogi todos los datos de station_id
  for i in range(0, len(df_n_station), 5):  # Bucle, recorri todos los datos de esta estacion de 5 en 5
    chunk =  df_n_station.iloc[i:i+5].copy()  # Cogi las 5 filas siguientes
    if len(chunk) < 5:  # Si el numero de filas no es múltiplo de 5, ignoramos las ultimas restantes que queden solas. Es decir, no las añadimos a nuestra base de datos.
      continue
    new_lin = chunk.iloc[4].copy()
    new_lin['ctx-4'] = chunk.iloc[0]['percentage_docks_available']
    new_lin['ctx-3'] = chunk.iloc[1]['percentage_docks_available']
    new_lin['ctx-2'] = chunk.iloc[2]['percentage_docks_available']
    new_lin['ctx-1'] = chunk.iloc[3]['percentage_docks_available']
    result_list.append(new_lin)

# Genere un a lista y los converti en Dataframe
result_df = pd.DataFrame(result_list)
result_df.head()
```

Ya tengo la base de datos procesada, solo me hace falta dividir en X e Y, siendo **(X)** el 'metadata_sample_submission_2024.csv' y **(Y)** los valores a predecir por el modelo.

Voy a ordenar por fecha para hacer los splits de Train-Validation-Test en el orden natural de los datos y evitar data linking que Pere menciono en el video.

```python
result_df_ordenado = result_df.sort_values(by=['year', 'month', 'day', 'hour'], ascending=[True, True, True, True])
```

```python
X = result_df_ordenado.drop(columns=['percentage_docks_available']) # Esto refleja la informacion de todos los Contextos con la Fecha.
Y = result_df_ordenado['percentage_docks_available'] # Es la informacion de la Proporcion de muelles disponibles
```

Divido el Train-Val-Test, los datos ordenados por fecha

```python
import  numpy as np
```

```python
X_array = np.array(X)
Y_array = np.array(Y)
```

```python
# Get the total number of samples
n_samples = len(X_array)

# Define the split sizes
train_size = int(0.8 * n_samples)  # 80% for training
val_size = n_samples - train_size  # Remaining for validation
```

```python
# Split the data without shuffling
X_train = X_array[:train_size]
Y_train = Y_array[:train_size]

X_val = X_array[train_size:]
Y_val = Y_array[train_size:]

print("Training set size:", len(X_train))
print("Validation set size:", len(X_val))
```

Obtuve el X_test del archivo 'metadata_sample_submission_2024.csv'

```python
filepath_metadata = '/content/drive/MyDrive/University of Barcelona/Capstone_Project_2024/metadata_sample_submission_2024.csv'
df_metadata = pd.read_csv(filepath_metadata, sep=',')
df_metadata
```

Añadimos la columna del año 2024

```python
df_metadata['year'] = 2024
X_test = df_metadata[['station_id', 'year', 'month', 'day', 'hour', 'ctx-4', 'ctx-3', 'ctx-2', 'ctx-1']]
X_test
```

```python
X_test = np.array(X_test)
```

Ahora voy a probar el modelo de regresión propuesto por Pere en su correo.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Initialize the model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, Y_train)

# Predict on the validation set
Y_val_pred = model.predict(X_val)

# Evaluate the model on the validation set
mse_val = mean_squared_error(Y_val, Y_val_pred)
r2_val = r2_score(Y_val, Y_val_pred)

print("Validation MSE:", mse_val)
print("Validation R^2:", r2_val)
```

Ya que le modelo está entrenado, ahora voy a calcular las predicciones para X_test de kaggle.

```python
# Predict on the test set
Y_test_pred = model.predict(X_test)
```
