```python
import py7zr

def unzip(file_path, output_dir):
    with py7zr.SevenZipFile(file_path, mode='r') as z:
        z.extractall(output_dir)
    print(f"{file_path}")

i2m = list(zip(range(1, 13), ['Gener', 'Febrer', 'Marc', 'Abril', 'Maig', 'Juny', 'Juliol', 'Agost', 'Setembre', 'Octubre', 'Novembre', 'Desembre']))
for year in [2023, 2022, 2021, 2020]:
    for month, month_name in i2m:
        zip_file = f'{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.7z'
        output_directory = 'Files'
        unzip(zip_file, output_directory)
```

```python
from collections import Counter
counter_ids = Counter(red_stations_df['station_id'])
list_ids = list(counter_ids.keys())
```

```python
list_ids
```

```python
result_list = []
for station_id in list_ids:
    df_n_station = sorted_df[sorted_df['station_id'] == station_id]
    for i in range(0, len(df_n_station), 5):
        chunk =  df_n_station.iloc[i:i+5].copy()
        if len(chunk) < 5:    # si el numero de filas no es múltiplo de 5, ignoramos las ultimas restantes qeu qeuden solas. Es decir, no las añadimos a nuestra base de datos
            continue
        new_lin = chunk.iloc[4].copy()
        new_lin['ctx-4'] = chunk.iloc[0]['percentage_docks_available']
        new_lin['ctx-3'] = chunk.iloc[1]['percentage_docks_available']
        new_lin['ctx-2'] = chunk.iloc[2]['percentage_docks_available']
        new_lin['ctx-1'] = chunk.iloc[3]['percentage_docks_available']
        result_list.append(new_lin)

result_df = pd.DataFrame(result_list)
```
